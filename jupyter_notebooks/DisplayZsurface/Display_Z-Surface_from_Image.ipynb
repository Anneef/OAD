```python
import os
import sys
sys.path.append(r'modules')
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import bftools as bf
```

```python
# define the desired options
writecsv = True
separator = '\t'
savefigure = True
saveformat = 'png'
surface = True

# specify the image filename
filename = r'c:\images\384well_DAPI.czi'
```

```python
bfpackage = r'modules\bioformats_package_5_1_10.jar'
bf.set_bfpath(bfpackage)

# create plane info from CZI image file and write CSV file
planetable, filenamecsv, MetaInfo = bf.get_planetable(filename, writecsv=writecsv, separator=separator, showinfo=True)
bf.kill_jvm()
```

```python
# show the dataframe
planetable[:10]
```

```python
# define name for figure to be saved (only the 1st figure will be saved)
figuresavename = os.path.splitext(filenamecsv)[0] + '_XYZ-Pos' + '.' + saveformat

# display the XYZ positions
fig1, fig2 = bf.scatterplot(planetable,
                            ImageID=0,
                            T=0,
                            CH=0,
                            Z=0,
                            size=250,
                            savefigure=savefigure,
                            figsavename=figuresavename,
                            showsurface=surface)
```
